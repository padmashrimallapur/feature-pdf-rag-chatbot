```python
from google.colab import files
uploaded = files.upload()
pdf_name= list(uploaded.keys())[0]
```

```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader(pdf_name)
document= loader.load()

print("pages :", len(document))
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_spliter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=100

)
docs= text_spliter.split_documents(document)
print('chunk :', len(docs))
```

```python
from  langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS

embeddings = OpenAIEmbeddings()
vectorstores= FAISS.from_documents(docs, embeddings)


```

```python
from langchain_openai import ChatOpenAI

llm=ChatOpenAI(
    model="gpt-3.5-turbo",
    temperature=0
)
```

```python
from langchain.chains import RetrievalQA

qa_chain=RetrievalQA.from_chain_type(
    llm=llm,
    retriever=vectorstores.as_retriever()
)
```

```python
qa_chain.run("What is the PDF about?")
```

```python
qa_chain.run("Suggest some breakfast recipies")
```

```python
qa_chain.run("Suggest some lunch recipies")
```
